## Combine Patches results

````python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
import cv2
````

````python
'''
get image names
'''
def read_image_list(path):
    f = open(path)
    lines = f.readlines()
    
    # remove .jpg\n
    lines = [i.split('.')[0] for i in lines]
    
    return lines
````

````python
'''
read txt predict files
'''
def predict(path, patch_size = 416, class_num = 5):
    
    # if file not exisit => no box
    try:
        f = open(path)
        lines = f.readlines()

        # remove .jpg\n
        lines = [i[:-1] for i in lines]

        # collect predictions
        # ex:
        # 3 0.25601 0.703125 0.0552885 0.0600962 0.316098
        # 3 0.121394 0.701923 0.09375 0.105769 0.599048
        predictions = []

        for box in lines:
            prediction = [] # ex : '3 0.25601 0.703125 0.0552885 0.0600962'
            for idx, attr in enumerate(box.split(' ')):
                if idx == 0:
                    prediction.append(int(attr))
                else:
                    prediction.append(float(attr))

            predictions.append(prediction)

        # parse predictions
        msk = np.zeros((patch_size, patch_size, class_num), dtype=float)

        for prd in predictions:

            x, y, w, h = prd[1:5]
            x, y, w, h = round(x * patch_size), round(y * patch_size), round(w * patch_size), round(h * patch_size)

            # print('class {}'.format(class_names.index(prd[0])))
            # print((round(w), round(h)))
            msk[y-h//2:y+h//2, x-w//2:x+w//2, prd[0]] = float(prd[5])

        return msk
    
    # empty mask
    except:
        return np.zeros((patch_size, patch_size, class_num), dtype=float)
````

````python
'''
concat patches inference results  
'''
def concat_inference(path, names):
    '''
    tiling strategy:
    patch size = 416*416
    shift = 208
    will be divided into = 18*24 = 432 patches (per img)
    '''
    
    patch_size = 416
    shift_size = 208
    # define patch numbers
    patch_y = 17
    patch_x = 23
    # define whole image
    whole_y = 3684
    whole_x = 4912
    
    for idx, n in tqdm(enumerate(names)):  
        # prediction for each whole image
        whole_mask = np.zeros((3684, 4912, 5), 'float')   
        
        for y in range(patch_y):
            for x in range(patch_x):
                # model inference part
                tile_path = os.path.join(path, '{}[{}][{}].txt'.format(n, y, x))
                # predict
                msk = predict(tile_path)
                
                # fill in the mask part
                if x == patch_x-1 and y == patch_y-1:
                    whole_mask[-patch_size:, -patch_size:, ...] += msk
                elif x == patch_x-1:
                    whole_mask[shift_size*y: shift_size*y+patch_size, -patch_size:, ...] += msk
                elif y == patch_y-1:
                    whole_mask[-patch_size:, shift_size*x: patch_size+shift_size*x, ...] += msk
                else:
                    whole_mask[shift_size*y: shift_size*y+patch_size, shift_size*x: patch_size+shift_size*x, ...] += msk
                
        
        # set confidence higher than one to one
        whole_mask = np.where(whole_mask > 1, 1.0, whole_mask)
        
        # save result
        for i in range(5):
            save_mask = np.stack([whole_mask[..., i], whole_mask[..., i], whole_mask[..., i]], axis=-1)*255
            cv2.imwrite('predictions/{}_predict{}.png'.format(n, i), save_mask)
            
        
    print('finish inference')
````

````python
'''
inference on patches
'''
def patches_inference(path, names):    
    # iterate through all the patches to do inference
    print('start inference')

    print('images num : {}'.format(len(names)))

    concat_inference(path, names)
    
    print("Done.")
````
