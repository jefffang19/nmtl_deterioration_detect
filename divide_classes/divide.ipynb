```python
# divide mixed labels into each classes a file
```

```python
import glob
import os
```

```python
yolo = sorted(glob.glob("../yolo/image*.txt"))
yolo2 = sorted(glob.glob("../yolo2/image*.txt"))
```

```python
len(yolo), len(yolo2)
```

```python
yolo
```

# iterate

```python
for yolo_file in yolo:
    # read mixed labels
    classes = [[], [], [], [], []]
    
    f = open(yolo_file)

    for i in f.readlines():
        classes[int(i[0])].append(i[:-1])
        
    f.close()
    
    # write to label file
    for i in range(5):
        f = open(os.path.join("yolo", str(i), yolo_file.split("/")[-1]), "w")
        
        for line in classes[i]:
            f.write("0{}\n".format(line[1:]))
        
        f.close()
```
